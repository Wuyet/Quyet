Bài 13: Viết hàm trả về tổng bình phương các số từ 1 đến n

Bài 14:  Viết hàm trả về hệ số nhị thức C(n, k).

```python
n,k= map(int, input().split())
def C(n,k):
    s=1
    h=1
    m=1
    for i in range(1,n+1):
        s*=i
    for e in range(1,k+1):
        h*=e
    for g in range(1, n-k+1):
        m*=g
    return s//(h*m)
print(C(n,k))
```

Bài 15: Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a, b, c dưới dạng danh sách [max, min]. (Không dùng tuple hay
dictionary) 

```python
a,b,c= map(int, input().split())
def max_min(a,b,c):
    d= max(a,b,c)
    e= min(a,b,c)
    return [d,e]
print(max_min(a,b,c))
```
