Bài 3: Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. 

```python
n =int(input())
if n>0 and n&(n-1) ==0: #sử dụng bitwise AND
    print("True")
else:
    print("False")
```

Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 


```python
x= int(input())
if 0 < x <= 50:
    tong= x*1500
elif 51<= x <= 100:
    tong= 50*1500 + (x-50)*2000
else:
    tong = 50*1500 + 50*2000 + (x-100)*2500
print(tong)
```
