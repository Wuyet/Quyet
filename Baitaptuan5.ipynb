Bài 10: Viết hàm kiểm tra 2 từ có đẳng cấu không?


```python
a,b= map(str, input().split())
map_a_to_b= {}
map_b_to_a= {}
def dang_cau(a,b):
    for i in range(len(a)):
        if a[i] in map_a_to_b:
            if map_a_to_b[a[i]] != b[i]:
                return False
        else:
            map_a_to_b[a[i]]= b[i] # gán keys và values vào dict
        if b[i] in map_b_to_a:
            if map_b_to_a[b[i]] != a[i]:
                return False
        else:
            map_b_to_a[b[i]] = a[i]
print(dang_cau(a,b))
        
```
