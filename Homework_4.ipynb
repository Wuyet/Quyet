Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n= int(input())
while not n==1:
    if n%2!=0:
        print('False')
        break
    else:
        n/=2
else:
    print("True")
```

Bài 27:Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắn

```python
a= input()
dem=0
for i in a:
    if i.isalpha():
        dem+=1
print(dem)
```

Bài 31: Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in
thường, bao nhiêu ký tự số

```python
a= input()
in_hoa= 0
in_thuong=0
so= 0
for i in a:
    if i.isdigit():
        so+=1
        continue
    if i.isupper():
        in_hoa+=1
        continue
    elif i.islower():
        in_thuong+=1
        continue
print(in_hoa, in_thuong, so)
```
