Bài 8: Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n_str= input()
n = int(n_str)
while str(n) != n_str[::-1]:
    rev_n= int(n_str[::-1])
    n= n+rev_n
    print(f"Gia tri doi xung la {n}")
    break
```

Bài 13: Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a,b= map(int,input().split())
list_a= []
list_b= []
while not (a>0 and b>0):
    a,b= map(int,input())
for i in range (1,a):
    if a%i ==0:
        list_a.append(i)
for e in range (1,b):
    if b%e ==0:
        list_b.append(e)
if sum(list_a) == b or sum(list_b)== a:
    print("True")
else:
    print("False")

```
