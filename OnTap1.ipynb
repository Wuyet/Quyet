Bài 1: Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
import math
car_type= input()
start_hour, km, wait_minutes= map(float, input().split())
def smart_taxi_care(car_type, start_hour, km, wait_minutes):
    #Phí từng loại xe
    if car_type =='xe 4 cho':
        if 0<= km <= 0.8:
            tien_km= 11000
        elif 0.8< km <=20:
            tien_km= 11000 + (km-0.8)*13000
        else:
            tien_km= 11000 + (20-0.8)*13000 + (km-20)*10500
    else:
        if 0<= km <= 0.8:
            tien_km= 12000
        elif 0.8< km <=20:
            tien_km= 12000 + (km-0.8)*14000
        else:
            tien_km= 12000 + (20-0.8)*14000 + (km-20)*12000

    # phí chờ 
    if wait_minutes >5:
        tien_cho= (math.ceil((wait_minutes-5)/5))*2000
    else:
        tien_cho=0  

    # Phụ thu ban đêm
    tien_tam_tinh= tien_km + tien_cho
    if 22<= start_hour <24 or 0<= start_hour <=5:
        tong = tien_tam_tinh* (1+0.2)  
    else:
        tong= tien_tam_tinh
        if km >30:
            tong = tong * (1-0.1)
    if tong< 50000 and km>5:
        tong= 50000
    return tong
print( smart_taxi_care(car_type, start_hour, km, wait_minutes))

                                
            
            
```

```python
n,k= map(int, input().split())
def C(n,k):
    s=1
    h=1
    m=1
    for i in range(1,n+1):
        s*=i
    for e in range(1,k+1):
        h*=e
    for g in range(1, n-k+1):
        m*=g
    return s//(h*m)
print(C(n,k))
```

Bài 27: Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
- Đầu vào: các số thực a, b, c.
- Đầu ra: 0, 1, hoặc 2. Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').
- Quy tắc:
- Nếu a != 0, dùng biệt thức D = b^2 - 4ac.
- Nếu a == 0 và b != 0, đó là phương trình bậc nhất với đúng một nghiệm thực.
- Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm.

```python
a,b,c= map(float, input().split())
def quadratic_roots_count(a, b, c):
    if a==0:
        if b==0:
            if c==0:
                return 'infinite'
        else:
            return '1'
    else:
        delta= b**2 -4*a*c
        if delta <0:
            return '0'
        elif delta==0:
            return '1'
        else:
            return '2'
print(quadratic_roots_count(a, b, c))
```

Bài 30: Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu
chấm động, ngược lại trả về 'neither'.
- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
s= input()
def detect_number_type(s):
    # xét s nếu s là chuỗi rỗng
    if not s:
        return 'neither'
    
    if s[0] in'+-':
        s= s[1:]
    
    if not s:
        return 'neither'
    
    if s.isdigit():
        return 'int'
    
    if s.count('.') ==1:
        left, right= s.split('.')
        if left.isdigit() and right.isdigit():
            return 'float'
    return 'neither'
print(detect_number_type(s))
        
```

Bài 38: Viết hàm trả về chuỗi biểu diễn hình chữ nhật N x M bằng ký tự '', các hàng cách nhau bởi ký tự xuống dòng.
Ví dụ, n=2, m=3

```python
n,m = map(int, input().split())
def rectangle(n, m):
    result = ""
    for i in range(n):
        result += "*" * m
        if i != n - 1:  # tránh xuống dòng ở dòng cuối
            result += "\n"
    return result
print(rectangle(n, m))
```
