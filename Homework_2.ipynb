Bài 3: Nhập vào 3 cạnh của tam giác. Kiem tra 3 cạnh có tạo thành tam giác vuông, cân, đều 

```python
a,b,c= map(float, input("Nhap vao 3 canh cua tam giac:").split())
if a>0 and b>0 and c>0:
    if a+b>c and a+c>b and b+c>a:
        if a==b==c:
            print("Tam giac deu")
        elif a==b or a==c or b==c:
            print("Tam giac can")
        elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
            print("Tam giac vuong")
        else:
            print("Tam giac thuong")
        chu_vi = a+b+c
        s= chu_vi/2
        dien_tich = (s*(s - a)*(s - b)* (s - c))**0.5
        print(f"Dien tich tam giac la: {dien_tich}")
        print(f"Chu vi cua tam giac la: {chu_vi}")
    else:
        print("Khong la 3 canh cua mot tam giac")
else:
    print("Khong la 3 canh cua mot tam giac")
```

Bài 8: Giải hệ phương trình tuyến tính

```python
a,b,c,d= map(float, input().split())
m,n = map(float, input().split())
det= a*d-b*c
det_x= m*d-b*n
det_y= a*n-m*c
if det!=0:
    x= det_x/det
    y= det_y/det
    print(x,y)
else:
    if det_x==det_y==0:
        print("Vo so nghiem")
    else:
        print("Vo nghiem")
```
