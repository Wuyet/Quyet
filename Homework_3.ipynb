Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a,b= map(int,input().split())
list_a=[]
list_b=[]
list=[]
for i in range (1,a+1):
    if a%i==0:
        list_a.append(i)
for e in range (1,b+1):
    if b%e==0:
        list_b.append(e)
for c in list_a:
    for d in list_b:
        if c==d:
            list.append(c)
print(list)
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 

```python
n= int(input())
chan =0
le =0
for i in str(n):
    if int(i)!=0:
        if int(i)%2==0:
            chan+=1
        else:
            le+=1
    else:
        continue
print(chan, le)
```

```python
n= int(input())
chan=0
le=0
while n>0:
    digit= n%10
    if digit==0:
        n//=10
        continue # bắt đầu vòng lặp mới không xét các câu lệnh dưới
    
    if digit%2==0:
        chan+=1
    else:
        le+=1
    n//=10 
print(chan, le)
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
a= int(input())
f1, f2 = 1,1
if a< 1:
    print('Khong co só fibonacci')
else:
    while f2 <a:
        f1,f2= f2,f1+f2 #giá trị của f1 và f2 cập nhập cùng lúc, thực hiện vế bên phải trước
print(f1)
```
